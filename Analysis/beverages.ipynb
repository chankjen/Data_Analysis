```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

# Load the Excel file
beaverages_df = pd.read_excel('beaverages.xlsx')

# Display the first 5 rows
print("Beaverages Dataset:")
print(beaverages_df.head())
```

```python
# Calculate average price
average_price = beaverages_df['Price ($)'].mean()
print(f"\nAverage Price of Beaverages: ${average_price:.2f}")
```

```python
# Find the most popular beverage
most_popular = beaverages_df.loc[beaverages_df['Popularity (1-10)'].idxmax(), 'Beverage']
print(f"Most Popular Beaverage: {most_popular}")
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(beaverages_df['Price ($)'], bins=5, kde=True, color='skyblue')
plt.title('Distribution of Beaverage Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.pie(beaverages_df['Popularity (1-10)'], labels=beaverages_df['Beverage'], autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'violet'])
plt.title('Popularity of Beverages')
plt.show()
```
